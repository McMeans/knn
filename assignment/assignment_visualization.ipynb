````python
# PART 1
import pandas as pd

# load df
df = pd.read_csv('./data/college_completion.csv')
````

````python
# PART 2

# print dimensions
print(f"Dimensions: {df.shape}")
print(f"Observations: {df.shape[0]}")
print(f"Variables: {df.shape[1]}")

df.head()
````

````python
# PART 3

# cross tabulate `control` and `level`
cross_tab = pd.crosstab(df['control'], df['level'])
print(cross_tab)
````

The number of 2-year and 4-year private for-profit schools are close. There are significanlty more 4-year private non-profit schools than 2-year. Finally, there are a larger amount of public 2-year schools present than 4-year.

````python
# PART 4
import seaborn as sns
import matplotlib.pyplot as plt

# plot histogram
sns.histplot(df['grad_100_value'], kde=False)
plt.title('Histogram of grad_100_value')
plt.show()

# plot kernel density
sns.kdeplot(df['grad_100_value'], fill=True)
plt.title('Kernel Density Plot of grad_100_value')
plt.show()

# plot boxplot
sns.boxplot(df['grad_100_value'])
plt.title('Boxplot of grad_100_value')
plt.show()

# statistical description
print(df['grad_100_value'].describe())
````

````python
# PART 5

# control grouped kernel density plot
sns.kdeplot(data=df, x='grad_100_value', hue='control', fill=True)
plt.title('Grouped Kernel Density Plot of grad_100_value by control')
plt.show()

# level grouped kernel density plot
sns.kdeplot(data=df, x='grad_100_value', hue='level', fill=True)
plt.title('Grouped Kernel Density Plot of grad_100_value by level')
plt.show()
````

There seems to be a lot of visual consistency between the 2-year and public portions of their respective plots. This includes their density peaks, range of values, and shape. All other variables are more spread out, and their peaks are not as extreme.

````python
# descriptions of 'grad_100_value' by 'level'
grouped_by_level = df.groupby('level')['grad_100_value'].describe()
print("Grouped by level:")
print(grouped_by_level)

# descriptions of 'grad_100_value' by 'control'
grouped_by_control = df.groupby('control')['grad_100_value'].describe()
print("\n\nGrouped by control:")
print(grouped_by_control)

# descriptions of 'grad_100_value' by 'level' and 'control'
grouped_by_lc = df.groupby(['level', 'control'])['grad_100_value'].describe()
print("\n\nGrouped by level and control:")
print(grouped_by_control)
````

Based on these summaries, private 4-year non-for-profit schools look to have the best graduation rates.

````python
# PART 6

# create column
df['levelXcontrol'] = df['level'] + ', ' + df['control']

# plot data
sns.kdeplot(data=df, x='grad_100_value', hue='levelXcontrol', fill=True)
plt.title('Grouped Kernel Density Plot of grad_100_value by level and control')
plt.show()
````

Based on this grouped plot, private 4-year not-for-profit schools once again look to be the best in graduation rate.

````python
# PART 7

# kernel density plot for aid_value
sns.kdeplot(data=df, x='aid_value', fill=True)
plt.title('Kernel Density Plot of aid_value')
plt.show()

# grouped by level and control
sns.kdeplot(data=df, x='aid_value', hue='levelXcontrol', fill=True)
plt.title('Grouped Kernel Density Plot of aid_value by level and control')
plt.show()
````

The shape of the graph shows that 4-year private not-for-profit schools gives larger amounts of aid than any other type. Most of the other types of schools have peaks around $5,000 with a max of about $20,000, while this type peaks just below $15,000 and has a max of almost $50,000.

````python
# descriptions of aid_value by level and control
grouped_by_level_control = df.groupby('levelXcontrol')['aid_value'].describe()
print("Grouped by level and control:")
print(grouped_by_level_control)
````

````python
# PART 8

# scatterplot of grad_100_value by aid_value
sns.scatterplot(data=df, x='aid_value', y='grad_100_value')
plt.title('Scatterplot of grad_100_value by aid_value')
plt.xlabel('Aid Value')
plt.ylabel('Graduation Rate')
plt.show()
````

There is a positive correlation here. Based on the plot, there could be reason to believe an increase in aid can increase graudation rate. Though some fall below the trend (and some don't need aid to have high graduation rates), the general concept is prevalent.

````python
# scatterplot grouped by level and control
sns.scatterplot(data=df, x='aid_value', y='grad_100_value', hue='levelXcontrol')
plt.title('Scatterplot of grad_100_value by aid_value, grouped by level and control')
plt.xlabel('Aid Value')
plt.ylabel('Graduation Rate')
plt.legend(title='Level and Control')
plt.show()
````

The biggest player in this linear correlation are the 4-year private not-for-profit schools. Most of the other groups don't give large enough aid to show their graduation rates improve. This group has enough varied data to show this trend.
